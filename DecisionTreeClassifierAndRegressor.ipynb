# Node Class

```python
class Node:
    def __init__(self,feature_index=None, threshold=None, left=None, right=None, info_gain=None, value=None):
        #for decision nodes
        self.feature_index=feature_index
        self.threshold=threshold
        self.left=left
        self.right=right
        self.info_gain=info_gain
        
        #foe leaf nodes
        self.value=value
```

# DecisionTree model

```python
class DecisionTree():
    def __init__(self,criterion,min_samples=2, max_depth=2):
        #initialsing root
        self.root=None
        #stopping condition
        self.criterion=criterion
        self.min_samples=min_samples
        self.max_depth=max_depth
        self.tree_type=None
    
    def build_tree(self,dataset,curr_depth=0):
        x,y=dataset[:,:-1], dataset[:,-1]
        num_samples, num_features=np.shape(x)
        if num_samples>=self.min_samples and curr_depth<=self.max_depth:
            best_split=self.get_best_split(dataset,num_samples, num_features)
#             print("best",best_split["info_gain"])
            if(best_split["info_gain"]>0):
                left_tree=self.build_tree(best_split["left_dataset"], curr_depth+1)
                right_tree=self.build_tree(best_split["right_dataset"],curr_depth+1)
                return Node(best_split["feature_index"],best_split['threshold'],left_tree,right_tree, best_split["info_gain"])
        
        #leaf_node
        leaf_val=self.get_leaf_value(y)
        return Node(value=leaf_val)
    
    def get_best_split(self,dataset, num_samples, num_features):
        max_info_gain=-float('inf')
        split={}
        for feature_index in range(num_features):
            col_x=dataset[:,feature_index]
            threshold_values=np.unique(col_x)
            for threshold in threshold_values:
                left_dataset,right_dataset=self.split(dataset,feature_index, threshold)
                if(len(left_dataset)>0 and len(right_dataset)>0):
                    parent_y,left_y, right_y=dataset[:,-1], left_dataset[:,-1], right_dataset[:,-1]
                    if(self.criterion=="Information_gain"):
                        curr_info_gain=self.information_gain(parent_y,left_y,right_y)
                    else:
                        curr_info_gain=self.gini_index(parent_y,left_y,right_y)
                    if(curr_info_gain>max_info_gain):
                        max_info_gain=curr_info_gain
                        split["info_gain"]=curr_info_gain
                        split["left_dataset"]=left_dataset
                        split["right_dataset"]=right_dataset
                        split["feature_index"]=feature_index
                        split["threshold"]=threshold
        return split
    
    def split(self,dataset, feature_index, threshold):
        left_dataset=np.array([row for row in dataset if row[feature_index]<=threshold])
        right_dataset=np.array([row for row in dataset if row[feature_index]>threshold])
        return left_dataset, right_dataset
    
    def get_leaf_value(self,Y):
        if(self.tree_type=="Classification"):
            lt=list(Y)
            mode=max(lt,key=lt.count)
            return mode
        val=np.mean(Y)
        return val
    
    def information_gain(self,parent,left_y, right_y):
        frac_l=len(left_y)/len(parent)
        frac_r=len(right_y)/len(parent)
        if(self.tree_type=="Classification"):
            info_gain=self.entropy(parent)-(frac_l*self.entropy(left_y)+frac_r*self.entropy(right_y))
            return info_gain
        info_gain=np.var(parent)-(frac_l*np.var(left_y)+frac_r*np.var(right_y))
        return info_gain
    
    def entropy(self,y):
        #signma -pilog2(pi)
        lt=np.unique(y)
        entropy=0
        for val in lt:
            pr=len(y[y==val])/len(y)
#             print(pr,end=" ")
            entropy+=-1*pr*np.log2(pr)
        return entropy
    def gini(self,y):
        lt=np.unique(y)
        p=1
        for i in lt:
            p-=(len(y[y==i])/len(y))**2
        return p
    
    def gini_index(self,parent,left_y,right_y):
        frac_l=len(left_y)/len(parent)
        frac_r=len(right_y)/len(parent)
        info_gain=self.gini(parent)-(frac_l*self.gini(left_y)+frac_r*self.gini(right_y))
        return info_gain
        
    
    def fit(self,X,y):
        y=y.to_frame()
        dataset=np.concatenate((X,y),axis=1)
        y=y.to_numpy()
        m=type(y[0][y.shape[1]-1])
        if(np.issubdtype(np.float64, m)):
            self.tree_type="Regression"
        else:
            self.tree_type="Classification"
        self.root=self.build_tree(dataset)
        
    #Predicting is done at here
    
    def make_prediction(self,x,root):
        if(root.value!=None):
            return root.value
        feature_val=x[root.feature_index]
        if(feature_val<=root.threshold):
            return self.make_prediction(x,root.left)
        return self.make_prediction(x,root.right)
    
    def predict(self, X):
        X=X.to_numpy()
        predictions=[self.make_prediction(row,self.root) for row in X]
        return predictions
    
    #printing the tree at here
    def printer(self,root,level,string):
        spaces=2*level*" "
        if self.root!=root:
            print(spaces+string,end="")
        if root.value!=None:
            print("Class "+str(root.value))
            return
        
        print("?(X"+str(root.feature_index)+" <= "+str(root.threshold)+")")
        self.printer(root.left,level+1,"Y: ")
        self.printer(root.right,level+1, "N: ")
    
        
    
    def plot(self):
        if(self==None):
            print("No information to print")
        self.printer(self.root,0,"")
        
```
